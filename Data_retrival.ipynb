```python
import pandas as pd 
import numpy as np
```

```python
user_input = input()
```

```python
df = pd.read_csv('main.csv')
```

```python
data = df['Name'].tolist()
```

```python
if user_input in data:
    print("yes")
```

```python
count =0
dic={}
for i in range(len(data)):
    if user_input in data[i]:
        dic[count] = df.iloc[i]
        count+=1    
```

```python
output = pd.DataFrame.from_dict(dic).T
```

```python
output.head(2)
```

```python
final_df = output.sort_values(by=['Rating'], ascending=False)
```

```python
final_df.drop('Unnamed: 0',axis='columns', inplace=True)
```
